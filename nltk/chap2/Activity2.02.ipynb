```python
from nltk import word_tokenize
import re
from nltk.stem import WordNetLemmatizer
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
```

```python
text = ''
with open('data/text_corpus.txt') as f:
    text = f.read()
print(text)
```

```python
def clean_text(text):
    text = re.sub(r'([^\s\w]|_)+', ' ', text)
    tokens = word_tokenize(text)
    lower_tokens = [token.lower() for token in tokens]
    stop_words = stopwords.words('english')
    sw_removed = [token for token in lower_tokens if token not in stop_words]
    lemmatizer = WordNetLemmatizer()
    lemmas = [lemmatizer.lemmatize(lower_token) for lower_token in sw_removed]
    return lemmas
```

```python
cleaned_text = clean_text(text)
print(cleaned_text)
```

```python
def get_word_frequency(tokens):
    word_counter = Counter(tokens)
    return word_counter
```

```python
frequency = get_word_frequency(cleaned_text)
print(frequency.most_common(50))
```

```python
def generate_word_cloud(text):
    wordcloud = WordCloud(width=1000, height=1000, background_color='white', max_words=50, min_font_size=10).\
    generate(text)
    return wordcloud
```

```python
wc = generate_word_cloud(text)
plt.imshow(wc)
plt.show()
```
