```python
import spacy
from spacy import displacy
```

```python
nlp = spacy.load('pt_core_news_sm')
frase = nlp('No imposto profissional o justo paga mais e o injusto menos, sobre o mesmo rendimento.')
```

```python
displacy.render(frase, style='dep', jupyter=True)
```

```python
frase_ent = nlp('Jesus Cristo nasceu de Maria no natal')
displacy.render(frase_ent, style='ent', jupyter=True)
```
